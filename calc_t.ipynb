```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from operator import add
def t_test(mean_1, std_1, size_1, mean_2, std_2, size_2):
    return abs((mean_1 - mean_2) / sqrt((std_1**2/size_1) + (std_2**2/size_2)))

error600 = pd.read_csv("error600.csv")
del error600['ignore']
del error600['ignore.1']
times600 = pd.read_csv("times600.csv")
del times600['ignore']
del times600['ignore.1']
times600.columns = list(map(add,times600.columns.values,['_t']*len(times600.columns)))
frames = [error600,times600]
et600 = pd.concat(frames, axis=1)
et600 = et600[et600 != -1]
et600 = et600.dropna()

error600c = pd.read_csv("error600c.csv")
del error600c['ignore']
del error600c['ignore.1']
del error600c['ignore.2']
times600c = pd.read_csv("times600c.csv")
del times600c['ignore']
del times600c['ignore.1']
del times600c['ignore.2']
times600c.columns = list(map(add,times600c.columns.values,['_t']*len(times600c.columns)))
frames = [error600c,times600c]
et600c = pd.concat(frames, axis=1)
et600c = et600c[et600c != -1]
et600c = et600c.dropna()

error68 = pd.read_csv("error68.csv")
del error68['ignore']
del error68['ignore.1']
times68 = pd.read_csv("times68.csv")
del times68['ignore']
del times68['ignore.1']
times68.columns = list(map(add,times68.columns.values,['_t']*len(times68.columns)))
frames = [error68,times68]
et68 = pd.concat(frames, axis=1)
et68 = et68[et68 != -1]
et68 = et68.dropna()

error68c = pd.read_csv("error68c2.csv")
del error68c['ignore']
del error68c['ignore.1']
del error68c['ignore.2']
times68c = pd.read_csv("times68c2.csv")
del times68c['ignore']
del times68c['ignore.1']
del times68c['ignore.2']
times68c.columns = list(map(add,times68c.columns.values,['_t']*len(times68c.columns)))
frames = [error68c,times68c]
et68c = pd.concat(frames, axis=1)
et68c = et68c[et68c != -1]
et68c = et68c.dropna()
```
